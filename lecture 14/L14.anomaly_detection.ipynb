```python
def IQR_method(df,n,features, k=1.5):
    """
    Takes a dataframe and returns an index list corresponding to the observations
    containing more than n outliers according to the Tukey IQR method.
    """
    outlier_list = []

    for column in features:

        # 1st quartile (25%)
        Q1 = np.percentile(df[column], 25)
        # 3rd quartile (75%)
        Q3 = np.percentile(df[column],75)

        # Interquartile range (IQR)
        IQR = Q3 - Q1

        # outlier step
        outlier_step = k * IQR

        # Determining a list of indices of outliers
        outlier_list_column = df[(df[column] < Q1 - outlier_step) | (df[column] > Q3 + outlier_step )].index

        # appending the list of outliers
        outlier_list.extend(outlier_list_column)

    # selecting observations containing more than x outliers
    outlier_list = Counter(outlier_list)
    multiple_outliers = list( k for k, v in outlier_list.items() if v > n )

    # Calculate the number of records below and above lower and above bound value respectively
    df1 = df[df[column] < Q1 - outlier_step]
    df2 = df[df[column] > Q3 + outlier_step]

    print('Total number of outliers is:', df1.shape[0]+df2.shape[0])

    return multiple_outliers
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Detect anomalies via IQR method
outlier_indices = IQR_method(df, n=1, features=features, k=4.5)

# # Prepare a plotting DataFrame with an 'anomaly' flag
df_out = df[features].copy()
df_out['anomaly'] = 'normal'
df_out.loc[outlier_indices, 'anomaly'] = 'outlier'
```

```python
# Evaluate how well IQR outliers correspond to actual frauds
# Treat IQR outlier=1 as "predicted fraud", normal=0 as "predicted nonâ€‘fraud"
df_out['pred_iqr'] = 0
df_out.loc[outlier_indices, 'pred_iqr'] = 1

y_true = df['Class']
y_pred = df_out['pred_iqr']

print("=== IQR Outlier Detection vs. True Fraud Labels ===")
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=['nonâ€‘fraud','fraud']))

cm = confusion_matrix(y_true, y_pred)
cm_df_out = pd.DataFrame(cm,
                     index=['true_nonâ€‘fraud','true_fraud'],
                     columns=['pred_nonâ€‘fraud','pred_fraud'])
print("\nConfusion Matrix:")
print(cm_df_out)
```

```python
outlier_fraction = len(df[df['Class'] == 1])/float(len(df[df['Class'] == 0]))
outlier_fraction
```

```python
features = ['V10', 'V14', 'V4', 'V12', 'V11', 'V17']
X = df[features].values
y_true = df['Class'].values 
```

### Splitting the Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_true, test_size=0.3, random_state=42, stratify=y_true)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### EllipticEnvelope

```python
model = EllipticEnvelope(contamination=outlier_fraction)
model.fit(X_train_scaled)

y_pred = model.predict(X_test_scaled)
y_pred = np.where(y_pred == -1, 1, 0) # 1 = anomaly, 0 = normal
```

```python
print("Confusion Matrix (EllipticEnvelope):")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report (EllipticEnvelope):")
print(classification_report(y_test, y_pred, digits=4))
```

### Isolation Forest

```python
iso_forest = IsolationForest(contamination=outlier_fraction, random_state=42)
iso_forest.fit(X_train_scaled)

y_pred_iso = iso_forest.predict(X_test_scaled)
y_pred_iso = np.where(y_pred_iso == -1, 1, 0)  # 1 = anomaly, 0 = normal
```

```python
print("Confusion Matrix (IsolationForest):")
print(confusion_matrix(y_test, y_pred_iso))
print("\nClassification Report (IsolationForest):")
print(classification_report(y_test, y_pred_iso, digits=4))
```

## One-Class SVM

```python
ocsvm = OneClassSVM(kernel='rbf', nu=0.01, gamma='scale')
ocsvm.fit(X_train_scaled)

y_pred_svm = ocsvm.predict(X_test_scaled)
y_pred_svm = np.where(y_pred_svm == -1, 1, 0) # 1 = anomaly, 0 = normal
```

```python
print("Confusion Matrix (One-Class SVM):")
print(confusion_matrix(y_test, y_pred_svm))

print("\nClassification Report (One-Class SVM):")
print(classification_report(y_test, y_pred_svm, digits=4))
```

## DBSCAN +  KNN

```python
eps_target = 0.4 
dbscan = DBSCAN(eps=eps_target, min_samples=5)
dbscan.fit(X_train_scaled)

core_samples = X_train_scaled[dbscan.core_sample_indices_]
core_labels = dbscan.labels_[dbscan.core_sample_indices_]


print(f"Number of core samples: {len(core_samples)}")
print(f"Core labels: {set(core_labels)}")  
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(core_samples, core_labels)

y_dist, y_pred_idx = knn.kneighbors(X_test_scaled, n_neighbors=1)
new_y_pred = np.where(y_dist > eps_target, 1, 0).ravel()  # 1 = anomaly, 0 = normal
```

```python
print("Confusion Matrix (DBSCAN + KNN):")
print(confusion_matrix(y_test, new_y_pred))

print("\nClassification Report (DBSCAN + KNN):")
print(classification_report(y_test, new_y_pred, digits=4))
```

### Local Outlier Factor

#### Finding the `best n_neighbors` on the training data 

```python
neighbor_values = range(200, 260, 10)
contamination = outlier_fraction  

f1_scores = []
best_model = None

for n in neighbor_values:
    lof = LocalOutlierFactor(n_neighbors=n, contamination=contamination)
    y_pred_train = lof.fit_predict(X_train_scaled)
    y_pred_train_binary = [1 if val == -1 else 0 for val in y_pred_train]
    
    score = f1_score(y_train, y_pred_train_binary)
    f1_scores.append(score)
    print(f"n_neighbors = {n}, f1_score (train) = {score:.4f}")
```

### Select `the best n_neighbors`

```python
best_n = neighbor_values[np.argmax(f1_scores)]
lof = LocalOutlierFactor(n_neighbors=best_n, contamination=outlier_fraction)

y_pred_lof = lof.fit_predict(X_test)
y_pred_binary = [1 if val == -1 else 0 for val in y_pred_lof] # 1 = anomaly, 0 = normal
```

### Evaluation on Test Data

```python
print("Confusion Matrix (lof):")
print(confusion_matrix(y_test, y_pred_binary))
print("\nClassification Report (lof):")
print(classification_report(y_test, y_pred_binary))
```
